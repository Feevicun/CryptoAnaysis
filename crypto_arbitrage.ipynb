To prepare and clean your data for analysis, complete the following steps:

### Step 1: For the bitstamp DataFrame, replace or drop all `NaN`, or missing, values in the DataFrame.

```python


# Replace or drop all NaNs or missing values in the bitstamp DataFrame
bitstamp.dropna(inplace=True)

# Confirm if there are any remaining NaN values
print(bitstamp.isnull().sum())
```

### Step 2: Use the `str.replace` function to remove the dollar signs ($) from the values in the Close column.

```python
# Remove the dollar signs ($) from the Close column using str.replace
bitstamp['Close'] = bitstamp['Close'].str.replace('$', '')

# Display the first few rows of the DataFrame to confirm the changes
print(bitstamp.head())
```

### Step 3: Convert the data type of the Close column to a `float`.

```python
# Convert the data type of the Close column to float
bitstamp['Close'] = bitstamp['Close'].astype(float)

# Display the first few rows of the DataFrame to confirm the changes
print(bitstamp.head())
```

### Step 4: Review the data for duplicated values, and drop them if necessary.

```python
# Check for duplicate rows in the DataFrame
duplicate_rows = bitstamp[bitstamp.duplicated()]

# If there are duplicates, drop them
if not duplicate_rows.empty:
    bitstamp.drop_duplicates(inplace=True)
    print("Duplicate rows dropped.")
else:
    print("No duplicate rows found.")
```

### Step 5: Repeat Steps 1–4 for the coinbase DataFrame.

```python


# Step 1: Replace or drop all NaNs or missing values in the coinbase DataFrame
coinbase.dropna(inplace=True)

# Step 2: Remove the dollar signs ($) from the Close column using str.replace
coinbase['Close'] = coinbase['Close'].str.replace('$', '')

# Step 3: Convert the data type of the Close column to float
coinbase['Close'] = coinbase['Close'].astype(float)

# Step 4: Check for duplicate rows in the DataFrame and drop them if necessary
duplicate_rows_coinbase = coinbase[coinbase.duplicated()]
if not duplicate_rows_coinbase.empty:
    coinbase.drop_duplicates(inplace=True)
    print("Duplicate rows dropped for coinbase DataFrame.")
else:
    print("No duplicate rows found for coinbase DataFrame.")
```

#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python

# Calculate the arbitrage spread for the early date
arbitrage_spread_early = coinbase_one_day_early - bitstamp_one_day_early

# Generate summary statistics for the arbitrage spread (Early Date)
arbitrage_spread_early_positive = arbitrage_spread_early[arbitrage_spread_early > 0]
print("Summary statistics for the arbitrage spread (Early Date) where the spread is greater than zero:")
print(arbitrage_spread_early_positive.describe())

# Calculate the arbitrage spread for the middle date
arbitrage_spread_middle = coinbase_one_day_middle - bitstamp_one_day_middle

# Generate summary statistics for the arbitrage spread (Middle Date)
arbitrage_spread_middle_positive = arbitrage_spread_middle[arbitrage_spread_middle > 0]
print("\nSummary statistics for the arbitrage spread (Middle Date) where the spread is greater than zero:")
print(arbitrage_spread_middle_positive.describe())

# Calculate the arbitrage spread for the late date
arbitrage_spread_late = coinbase_one_day_late - bitstamp_one_day_late

# Generate summary statistics for the arbitrage spread (Late Date)
arbitrage_spread_late_positive = arbitrage_spread_late[arbitrage_spread_late > 0]
print("\nSummary statistics for the arbitrage spread (Late Date) where the spread is greater than zero:")
print(arbitrage_spread_late_positive.describe())
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python


# Calculate the spread returns for the early date
spread_return_early = arbitrage_spread_early_positive / bitstamp_one_day_early

# Review the spread return DataFrame for the early date
print("Spread return for the early date:")
print(spread_return_early)
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

#### 3. Для кожної з трьох дат ще більше звузіть свої торгові можливості. Для цього визначте, скільки разів ваші угоди з позитивною прибутковістю перевищують мінімальний поріг у 1%, необхідний для покриття ваших витрат.

```python


# Determine the number of times profitable trades exceed the 1% minimum threshold for the early date
profitable_trades_early = spread_return_early[spread_return_early > 0.01]

# Review the first five profitable trades for the early date
print("Profitable trades for the early date:")
print(profitable_trades_early.head())
```

#### 4. Generate the summary statistics of your spread returns that are greater than 1%. How do the average returns compare among the three dates?

#### 4. Згенеруйте підсумкову статистику прибутковості спреду, що перевищує 1%. Як порівняти середню прибутковість між трьома датами?

```python

# Generate summary statistics for the profitable trades (spread returns > 1%) for the early date
print("Summary statistics for profitable trades (spread returns > 1%) for the early date:")
print(profitable_trades_early.describe())
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

#### 5. Для кожної з трьох дат обчисліть потенційний прибуток у доларах на угоду. Для цього помножте спред, який перевищує 1%, на вартість того, що було придбано. Обов’язково видаліть усі відсутні значення з результуючого DataFrame.

```python
# Calculate the potential profit per trade in dollars for the early date
profit_early = profitable_trades_early * bitstamp_one_day_early

# Drop any missing values from the profit DataFrame
profit_per_trade_early = profit_early.dropna()

# View the early profit DataFrame
print("Profit per trade for the early date:")
print(profit_per_trade_early)


```
